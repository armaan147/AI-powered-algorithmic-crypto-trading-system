# 3. Plotting Regression

```python
def plot_regression(df, y_pred, height=500):
    """Plot regression analysis."""
    plt.figure(figsize=(12, height / 100))
    plt.plot(df.index, df['close'], label='BTC Price', color='blue')
    plt.plot(df.index, y_pred, label='Regression Line', color='red', linestyle='--')
    plt.title('BTC Price with Regression Line')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
%matplotlib inline
plot_regression(btc_df, y_pred)
```
